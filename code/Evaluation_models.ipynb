```python
from keras.models import load_model
import numpy as np
from tensorflow.keras.preprocessing import image
from IPython.display import Image
```

```python
model = load_model(r'C:\Users\Lenovo\Desktop\DRIVE_DROWSINESS\model VGG\VggNet.keras')
```

```python
def prepare_image(file_path):
    img = image.load_img(file_path, target_size=(64, 64))
    img_array = image.img_to_array(img)
    img_array = img_array / 255.0
    img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    return img_array_expanded_dims
```

```python
Image(filename=r"C:\Users\Lenovo\Desktop\DRIVE_DROWSINESS\evaluation\Open\Open_12.jpg", width=300,height=200) 
```

```python
image_path = r"C:\Users\Lenovo\Desktop\DRIVE_DROWSINESS\evaluation\Open\Open_12.jpg"
preprocessed_image = prepare_image(image_path)
prediction = model.predict(preprocessed_image)[0][0]

if prediction > 0.8:
    print(f"L'image est prédite comme 'open_eyes' avec une confiance de {round(prediction * 100, 2)}%.")
else:
    print(f"L'image est prédite comme 'close_eyes' avec une confiance de {round((1 - prediction) * 100, 2)}%.")
```
